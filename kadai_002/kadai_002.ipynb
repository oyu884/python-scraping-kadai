```python
import requests
from getpass import getpass
from pprint import pprint

# APIキー・駅名入力
api_key = getpass("Google Maps APIキーを入力: ")
station = input("駅名を入力: ")

# APIエンドポイント
GEOCODE_URL = "https://maps.googleapis.com/maps/api/geocode/json"
PLACES_URL = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# 駅名 → 緯度経度
geo = requests.get(GEOCODE_URL, params={
    "address": station,
    "key": api_key
}).json()

pprint(geo["status"])

loc = geo["results"][0]["geometry"]["location"]
lat, lng = loc["lat"], loc["lng"]

# 周辺レストラン取得（半径500m）
places = requests.get(PLACES_URL, params={
    "location": f"{lat},{lng}",
    "radius": 500,
    "type": "restaurant",
    "language": "ja",
    "key": api_key
}).json()

pprint(places["status"])

# 結果出力
for r in places.get("results", []):
    print(r.get("name"), r.get("rating"), r.get("vicinity"))
```
